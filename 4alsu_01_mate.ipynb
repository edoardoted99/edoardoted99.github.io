```python
import random
import numpy as np

def generate_vector():
    while True:
        values = [random.uniform(0.2, 1) for _ in range(12)]
        thirteenth_value = 10 - np.sum(values)
        if 0.2 <= thirteenth_value <= 1:
            values.append(thirteenth_value)
            random.shuffle(values)
            return values

array = generate_vector()
print(array)
print("Sum:", np.sum(array))
```
